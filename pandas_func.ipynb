**1. Importing pandas library**

```python
import pandas as pd
```

**2. Data Structures:**

**Series() -> to create 1-d structure of data using any iterable**

```python
x = pd.Series([1,2,3])
print(x)
print(type(x))

print()

y = pd.Series(('1', True, 'aman'))
print(y)
print(type(y))

print()

z = pd.Series({'name':['aman','raj','madhav'], 'marks':[23,34,12], 'grade':['A', 'O', 'C']})
print(z)
print()

#using with some add. parametrs
a = pd.Series([1,3,6,9,11], dtype=float, index=['a','b','c','d','e'])
print(a)
print()

print(x[2])
print(y[1])
print(z[2])
print(z['marks'])
```

**Dataframe() -> to create 2-d structure, we have to pass the single parameter(can be nested) in DataFrame() to create a 2-d fig.**

```python
x = pd.DataFrame([2,10,3])
print(x)
print(type(x))

print()

t1 = (1, 'aman', True)
t2 = (45, 'vikram', 'False')
y = pd.DataFrame(t1,t2)
print(y)

d = {'name':['raj', 'nikhil', 'manav'], 'age':[23, 21, 25], 'marks':[23, 56, 78]}
z = pd.DataFrame(d)
print(z)

print()

z2 = pd.DataFrame(d, columns=['name','marks'])
print(z2)

print()

z3 = pd.DataFrame(d, index=['a', 'b', 'c'])
print(z3)

print()

l = [[1,2,3], [3,4,5], [3,7,8]]
z4 = pd.DataFrame(l)
print(z4)

print()

t = ([1,2,3], [3,4,5], [3,7,8], ('a', 'b', 'c'))
z5 = pd.DataFrame(t)
print(z5)

print()

s1 = pd.Series([1,2,3,4,5])
s2 = pd.Series(['a','b','c','d','e'])
z6 = pd.DataFrame({'first':s1, 'second':s2})
print(z6)
```

**3. Arithmetic operations:**

```python
d = {'1st':{1,2,3,4,5}, '2nd':{4,5,6,1,2}}
print(d)

print()

d = {'1st':[1,2,3,4,5], '2nd':[4,5,6,1,2]}
print(d)

x = pd.DataFrame(d)
print(x)

x['sum'] = x['1st'] + x['2nd']
print(x)

print()

x['diff'] = x['1st'] - x['2nd']
print(x)

print()

x['prod'] = x['1st'] * x['2nd']
print(x)

print()
x['div'] = x['1st'] / x['2nd']
print(x)

print()
x['mod'] = x['1st'] % x['2nd']
print(x)

print()
x['power'] = x['1st'] ** x['2nd']
print(x)

print()
```

**4. insert()**

```python
d = {'1st':[1,2,3,4,5], '2nd':[4,5,6,1,2]}
print(d)

x = pd.DataFrame(d)
print(x)

print()

#insert into original dataframe
x.insert(1, '3rd', [10,20,30,40,50])
print(x)
```

**delete: by pop() and del**

```python
var1 = x.pop('3rd')
print(var1)
print(x)

print()

del x['2nd']
print(x)
```
